```python
# Functions

# Questions
# Write a function that takes a single vector input and returns the sum of the squares of its elements as its single output
def square_elems(a):
    b = a**2
    c = sum(b)
    return c

# Use the function to sum the squares of the numbers from 27 to 37 inclusive
v = np.arange(27, 38)
z = square_elems(v)
z

# Write a function that takes a single vector input and returns the mean, mode, and median as three separate variables

def mean_mode_med(a):
    mean = np.mean(a)
    mode = stats.mode(a)[0]
    median = np.median(a)
    return mean, mode, median

(mean, mode, median) = mean_mode_med(np.random.randint(1000))
print(mean, mode, median)
```

```python
# Simulating ordinary differential equations in MATLAB (python)

# Two methods for simulating the differential equation dx/dt = 2sin(t^2 / 10) with initial condition x(0) = 2
# Need to specify range of time across which to record all values of the variables
# Therefore need to create vector of time points
# Then set initial conditions

# First method 
# Loops through time to integrate up the differential equation

# GOING TO FINISH THIS TOMORROW (ON 5/20) SINCE I'M TIRED AND GOING TO SLEEP 

x0 = 2                                 # first value of x
t0 = 0                                 # first value of time
dt = 0.001                             # time step for simulations
tmax = 20                              # maximum time used
tvec = np.arange(t0, tmax, dt)         # vector of time points
xvec = np.zeros(tvec.shape)
xvec[0] = x0

for i, t in enumerate(tvec.shape):
    tval = tvec[i]
    dxdt = 2 * np.sin(0.1*tval*tval)
    xvec[i] = xvec[i - 1] + (dt * dxdt)


plt.plot(tvec, xvec)
```

```python
# Second method
# Uses functions and ODE solve (using scipy.integrate.solve_ivp)

x0 = np.array([2])                     # first value of x (MUST BE 1-D ARRAY (NOT INT!!))
t0 = 0                                 # first value of time
dt = 0.001                             # time step for simulations
tmax = 20                              # maximum time used
tvec = np.arange(t0, tmax, dt)         # vector of time points

def test_ode(t, y):
    dydt = 2*np.sin(0.1*t*t)
    return dydt
    
sol = integrate.solve_ivp(test_ode, [0, 20], x0, t_eval=tvec)

sol
plt.plot(sol.t, sol.y[0])
```

```python
# Solving coupled ODEs with multiple variables
# Example
# Solving system of equations:
# dx(t)/dt = v(t)
# dv(t)/dt = -(w^2)x(t)

tmax = 100
x0 = 1.0
v0 = 0.0
omega = 3
omega_sqr = omega**2
dt = 0.00001
tvector = np.arange(0, tmax, dt)
x1 = np.zeros(tvector.shape)
x2 = np.zeros(tvector.shape)
x1[0] = x0
x2[0] = v0

for i in range(2, len(tvector)):
    x1[i] = x1[i - 1] + x2[i - 1] * dt
    x2[i] = x2[i - 1] - omega_sqr * x1[i - 1] * dt

# Plotting obviously not right 

plt.plot(tvector, x1)
plt.plot(tvector, x2)
    
```

```python
# Solving ODEs with nested for loops
# See how changing parameters affects behavior of ODE
# Simulate dx/dt = 2sin(t^2/K), K between 1-20
# Goal is to find how value of x at time t = 3 depends on parameter K when x[0] = 0

dt = 0.001                               # time step for simulation
tmax = 3                                 # max time for simulation
t = np.arange(0, tmax, dt)               # vector of time points
kvals = np.arange(0.1, 20.1, 0.01)       # set of values used for K
xfinal = np.zeros(kvals.shape)           # to record x[3] for each K

t.shape

for simulation, b in enumerate(kvals):   # outer loop alters K
    k = kvals[simulation]                # select K for simulation
    x = np.zeros(t.shape)                # initialize x for each new K
    
    for i in range(t.shape[0]):                          # inner loop through time
        x[i] = x[i - 1] + dt*2*np.sin((t[i]*t[i])/k) # update x
    
    xfinal[simulation] = x[-1]           # record x[3] for this k
    
plt.plot(kvals, xfinal)
```
